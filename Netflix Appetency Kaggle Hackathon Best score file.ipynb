# Auto Ml Model

```python
from flaml import AutoML
```

```python
# Initialize an AutoML instance
automl = AutoML()
# Specify automl goal and constraint
automl_settings = {
    "time_budget": 1800,  # in seconds
    "metric": 'accuracy',
    "task": 'classification',
    "log_file_name": "iris.log",
}
```

```python
automl.fit(x,y,**automl_settings)
```

```python
automlpred=automl.predict(test_data)
```

```python
automl_output=pd.DataFrame({'id':df_test['id'],'target':automlpred})
```
